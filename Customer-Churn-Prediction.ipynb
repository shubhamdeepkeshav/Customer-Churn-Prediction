# **Printing Unique Values for Object Data Type Columns in a DataFrame**

```python
def printUniqueValues(df):
    for col in df:
        if df[col].dtype == 'object': # only print data type object
            print(f'{col}: {df[col].unique()}')
```

#**"Separating Features and Target Variable in DataFrame 'df'**

```python
X = df2.drop('Churn', axis=1)
y = df2['Churn']
```

# **Splitting Data into Training and Testing Sets Using train_test_split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

# **Displaying the Shapes of Training and Testing Feature Sets**

# **Importing TensorFlow and Keras Modules for Building Neural Networks**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
```

# **Building a Sequential Neural Network Model with Keras**

```python
model = Sequential([
        keras.layers.Dense(26, input_shape=(26,), activation='relu'), # 26 input, 26 neutron
        keras.layers.Dense(15, activation='relu'), # 15 neutron
        keras.layers.Dense(1, activation='sigmoid') # 1 output layer (binrary classification)

])
```

# **Displaying the Summary of the Neural Network Model**

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

# **Compiling the Neural Network Model with Adam Optimizer and Binary Crossentropy Loss**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
import numpy as np # Import numpy for array manipulation
import pandas as pd # Import pandas for DataFrame manipulation

# ... (rest of your model definition)

# Drop non-numerical columns from X_train
X_train = X_train.select_dtypes(include=['number'])

# Convert X_train and y_train to float32
X_train = np.array(X_train).astype('float32')
y_train = np.array(y_train).astype('float32')

model.fit(X_train, y_train, epochs=100)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
import numpy as np
import pandas as pd

# ... (rest of your model definition)

# Assuming X_test is a pandas DataFrame, select numerical columns
X_test = X_test.select_dtypes(include=['number'])

# Convert X_test and y_test to float32
X_test = np.array(X_test).astype('float32') # Convert X_test to float32
y_test = np.array(y_test).astype('float32') # Convert y_test to float32

model.evaluate(X_test, y_test)
```
